```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.image import load_img

img = np.array(load_img("../datasets/panda.jpg"))
```

```python
plt.imshow(img)
```

```python
generator = ImageDataGenerator(rotation_range=25, horizontal_flip=True, shear_range=.1, zoom_range=.3)
```

```python
flow = generator.flow(np.array([img]))
```

```python
batch = next(flow)
```

```python
batch.shape
```

```python
batch = next(flow)
for i in range(len(batch)):
    plt.figure()
    plt.imshow(batch[i].astype(np.uint8))
    plt.show()
```
