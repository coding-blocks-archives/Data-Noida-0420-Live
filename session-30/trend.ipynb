```python
import pandas as pd
import matplotlib.pyplot as plt

import numpy as np
```

```python
data = pd.read_csv("AAPL.csv")
```

```python
data
```

```python
data_high = data.High.values
```

```python
plt.plot(data_high)
```

```python
## moving average


window = 30
half = window // 2

out = []

for i in range(len(data_high)):
    if i < half:
        cut = data_high[: i + half]
    else:
        cut = data_high[i - half: i + half]
    out.append(np.mean(cut))
```

```python
plt.plot(out)
plt.plot(data_high)

plt.plot(data_high - out)
```
