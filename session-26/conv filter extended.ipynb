```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.image import load_img
```

```python
img = np.array(load_img("../datasets/mon.jpg", target_size=(400, 600)))
```

```python
gray = img.mean(axis=2)
```

```python
plt.imshow(gray, cmap="gray")
```

```python
plt.imshow(img, cmap="gray")
```

```python
def apply_filter(gray, fil):
    
    padded_img = np.pad(gray, 1)
    result = np.empty_like(gray)

    for row in range(result.shape[0]):
        for col in range(result.shape[1]):
            cut = padded_img[row:row+3, col:col+3]
            result[row, col] = np.abs(np.mean(fil * cut))
            
    return result
```

```python
filter_1 = np.array([[1, 0, -1],
                     [1, 0, -1],
                     [1, 0, -1]])

plt.imshow(apply_filter(gray, filter_1), cmap="gray")
```

```python
filter_1 = np.array([[1, 1, 1],
                     [0, 0, 0],
                     [-1, -1, -1]])

plt.imshow(apply_filter(gray, filter_1), cmap="gray")
```

```python
filter_1 = np.array([[1, 1, 1],
                     [1, 1, 1],
                     [1, 1, 1]])

plt.imshow(apply_filter(gray, filter_1), cmap="gray")
```
