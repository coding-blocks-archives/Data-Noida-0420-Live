```python
x = np.linspace(-10, 10, 100)
```

```python
plt.figure()
plt.plot(x, np.sin(x), label="sin")
plt.plot(x, np.cos(x), label="cos")
plt.plot(x, np.tan(x), label="tan")
plt.legend()
plt.ylim(-4, 4)
plt.show()
```

```python
theta = np.linspace(-np.pi, np.pi, 100)
rad = 10000
```

```python
x = rad * np.cos(theta)
y = rad * np.sin(theta)
```

```python
plt.figure(figsize=(6, 6))
plt.plot(x, y)
```

```python
theta = np.linspace(0, 40 * np.pi, 1000)
rad = np.linspace(0, 100, 1000)
```

```python
plt.figure(figsize=(6, 6))
plt.plot(x, y)
```
