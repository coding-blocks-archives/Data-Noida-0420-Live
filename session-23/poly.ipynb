```python
import tensorflow as tf

from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt

# Hide warnings
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
```

```python
noise = np.random.uniform(-50, 50, 10000)
```

```python
X = np.linspace(-20, 20, 10000)
```

```python
y = 3 * X * X + 5 * X + 10 + noise
```

```python
plt.scatter(X, y)
```

```python
X_mod = (X - X.mean())/X.std()
```

```python
Y_mod = (y - y.mean())/y.std()
```

```python
plt.scatter(X_mod, Y_mod)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X_mod, Y_mod, test_size=0.33, random_state=42)
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

model = Sequential()
model.add(Dense(units=4, activation="sigmoid", input_shape=(1,)))
model.add(Dense(units=1, activation="linear"))

model.compile(optimizer='adam', loss='mse')
```

```python

model.fit(X_train, y_train, batch_size=100, epochs=50, verbose=0)
```

```python
yp = model.predict(X_test).flatten()
```

```python
plt.scatter(X_train, y_train)
plt.scatter(X_test, yp)
```

```python
model.evaluate(X_test, y_test, verbose=0)
```
