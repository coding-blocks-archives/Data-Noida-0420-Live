```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
m, c = 4, 0
```

```python
X = np.linspace(0, 50, 100)

noise = np.random.randn(100) * 20
```

```python
y = m * X * X - 10 * X + noise
```

```python
plt.scatter(X, y)
```

```python
class CustomLinearReg:
    
    def __init__(self, lr=.1, iters=10):
        self.lr = lr
        self.iters = iters
    
    def fit(self, X, y):
        self.coef_ = np.random.random()
        
        errors = []
        for i in range(self.iters):
            error = self.gradient_decend(X, y)
            errors.append(error)
        
        return errors
    
    def gradient_decend(self, X, y):
        delta_coef, error = self.gradient(X, y)
#         print(delta_coef)
        self.coef_ += delta_coef
        return error
        
    def gradient(self, X, y):
        yh = self.hypo(X)
        dmsedm = np.mean(2 * (yh-y) * X)
        return -dmsedm * self.lr, np.mean((yh-y)**2)
        
    def hypo(self, X):
        return X * self.coef_
```

```python
for i in range(5, 15):

    model = CustomLinearReg(lr=.0001*i, iters=30)
    errors = model.fit(X, y)
    
    plt.figure()
    plt.title(str(.0001*i))
    plt.plot(errors)
    plt.show()
```
