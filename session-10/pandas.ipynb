```python
def gen_output(row):
    temp = row.copy()
    temp["output"] = row.sum()
    return temp

df.apply(gen_output, axis=1)
```

```python
def gen_double(row):
    row[:] = row * 2
    return row

df.apply(gen_double, axis=1)
```
